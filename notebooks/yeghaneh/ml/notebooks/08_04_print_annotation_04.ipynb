## Read the Data

```python
import pandas as pd
importVersion = '013'
```

```python
path= '../data/01_df_v{0}.pickle'.format(importVersion)# Put the path of the data in your local machine here, consider the letter "r" before the path
dfAstroNova = pd.read_pickle(path)
```

```python
dfAstroNova.head(5)
```

```python
texts=[]
for sen in dfAstroNova.sentences:
    texts +=sen
    
```

```python
len(texts)
```

```python
texts=texts[0:1000]  #choose part of data 
```

```python
str1 = ' '.join(texts)
```

```python
type(texts)
```

# Upload customized spaCy model provioded by our training data in prodigy

```python
path= '../data/Model_01'
nlp=spacy.load(path)
```

```python
doc=nlp(str1)
```

```python
options={"ents": ["TIME", "DATE","PARA","ASTR","LONG" ,"STAR","PLAN","NAME"],"colors":{"TIME":"CORAL","DATE":"TOMATO","PARA":"YELLOW","ASTR":"AQUA","LONG":"MAGENTA","STAR":"GREEN","PLAN":"LIME","NAME":"MEDIUM"}}
```

# Colorful Print of Entites  (labels)

```python
displacy.render(doc,style="ent",jupyter="True",options=options )
```
